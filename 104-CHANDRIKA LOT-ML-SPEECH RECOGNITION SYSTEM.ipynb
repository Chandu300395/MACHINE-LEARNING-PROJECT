```python
#Speech With Microphone 
import speech_recognition as sr
import pyttsx3
```

```python
def SpeakNow(command):
    voice=pyttsx3.init()
    voice.say(command)
    voice.runAndWait()
```

```python
r=sr.Recognizer()

with sr.Microphone() as source2:

    print("Silence Please")

    r.adjust_for_ambient_noise(source2,duration=1)

    print("Speack now Please......")

    audio2=r.listen(source2)

    texxt=r.recognize_google(audio2)

    texxt=texxt.lower()

    print("Did you say :- "+texxt)

    SpeakNow(texxt)
```
